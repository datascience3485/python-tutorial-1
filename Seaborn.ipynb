```python
import pandas as pd
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
print(sns.get_dataset_names())
```

```python
import matplotlib.pyplot as plt
```

```python
current_palette1=sns.color_palette()

sns.palplot(current_palette1)

plt.show()
```

```python
current_palette2=sns.color_palette()

sns.palplot(sns.color_palette("Greens"))
plt.show()
```

```python
current_palette3=sns.color_palette()
sns.palplot(sns.color_palette("BrBG",8))
plt.show()
```

```python
sns.distplot(tips['total_bill'])
plt.show()
```

```python
sns.distplot(tips['total_bill'],kde=False,hist=False)
plt.show()
```

```python
sns.distplot(tips['total_bill'],kde=True,hist=False)
plt.show()
```

```python
#jointplot
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips)
plt.show()
```

```python
sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.show()
```

```python
sns.set_style("ticks")
sns.pairplot(tips)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=tips)

plt.show()
```

```python
sns.scatterplot(x="day",y="total_bill",data=tips)

plt.show()
```

```python
sns.swarmplot(x="day",y="total_bill",data=tips,palette="husl")

plt.show()
```

```python
sns.boxplot(x=tips["total_bill"])
plt.show
```

```python
sns.boxplot(x="day",y="total_bill",data=tips)
plt.show()
```

```python
sns.barplot(x="day",y="total_bill",data=tips)
plt.show()
```

```python
sns.pointplot(x="time",y="total_bill",data=tips)
plt.show()
```

```python
sns.pointplot(x="time",y="total_bill",hue="smoker",data=tips)
plt.show()
```

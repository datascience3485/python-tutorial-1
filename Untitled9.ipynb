```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("Sample.xls")
```

```python
df.head()
```

```python
plt.scatter(x=df.Sales,y=df.Profit)

plt.xlabel("Sales")
plt.ylabel("Profit")
```

```python
sns.relplot(x='Sales',y='Profit',data=df)
plt.show()
```

```python
df["Order Priority"].value_counts()
```

```python
sns.relplot(x='Sales',y='Profit',data=df,hue='Order Priority')
plt.show()
```

```python
sns.relplot(x='Sales',y='Profit',data=df,hue='Order Priority',style='Order Priority')
```

```python
sns.relplot(x='Sales',y='Profit',data=df,hue='Order Priority',style='Ship Mode')
```

```python
sns.relplot(x='Sales',y='Profit',data=df,size='Discount')
```

```python
sns.relplot(x='Sales',y='Profit',data=df,size='Discount',sizes=(25,200))
```

```python
sns.catplot(x='Ship Mode',y='Profit',data=df)
```

```python
sns.catplot(x="Product Category",y="Shipping Cost",hue="Ship Mode",data=df,kind='box')
```

```python
sns.catplot(x="Product Category",y="Shipping Cost",col="Ship Mode",data=df,kind='box')
```

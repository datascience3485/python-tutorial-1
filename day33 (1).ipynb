```python
import pandas as pd
```

```python
data=pd.read_csv('gun-violence-data.csv')
data.head()
```

```python
data.info()
```

```python
data=data.iloc[:,:7]
data.head(10)
```

```python
data.drop(columns=['incident_id','city_or_county','address'],inplace=True)
data.head()
```

```python
data['date']=pd.to_datetime(data['date']) # convert date to date time format
data['year']= data['date'].dt.year # extract year
data['month']=data['date'].dt.day # extract day
data['day_of_week']= data['date'].dt.dayofweek # extract day of week
data['total_victim']= data['n_killed'] + data['n_injured'] # sum up numbers of deaths and injured
data.head()
```

```python
# Replace numeric day of week values with text

data['day_of_week'] = data['day_of_week'].map({0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'})
data.head()
```
